Using chr() (reverse of ord()).


````python
a = [56, 8, 103, 157, 22, 22, 24, 10, 100]

def convert_string(a):
  s = ''
  for i in a:
    s += chr(i)
  return s

convert_string(a)
````

````python
a = [56, 8, 103, 157, 22, 22, 24, 10, 100]

def convert_string(a):
  result = []
  for i in a:
    result.append(chr(i))
  return "".join(result)

convert_string(a)
````

#Task:
Convert a string to lowercase without using any inbuilt function like .lower().

````python
def custom_lower_case(s):
  result = ''
  for i in s:
      if ord(i) >= 65 and ord(i) <= 90:
          order = ord(i)
          order = order + 32
          result = result + chr(order)
      else:
          result  = result + i
  return result

custom_lower_case("HaRsHa")
````

* Splits the string by commas

* Returns the cleaned list of integers


>Expected output:

>(Big hint: .isdigit() won‚Äôt work for negatives ‚Äî think bigger.)

**2Ô∏è‚É£ MEDIUM**

>Write a function:  custom_lower(s)


that takes a string and returns a fully lowercased version, without using .lower().

Example:
```
Input: "ThiS Is A TeST"
Output: "this is a test"
```

You must:

* Use ASCII ranges (65‚Äì90)

* Only convert uppercase ‚Üí lowercase

* Leave all other characters unchanged

````python
def custom_lower(s):
  result = ''
  order = 0
  for i in s:
    if ord(i) >= 65 and ord(i) <= 90:
      order = ord(i)
      order += 32
      result += chr(order)
    else:
      result += i
  return result

custom_lower("ThiS Is A TeST")
````

**3Ô∏è‚É£ HARD**

>Modify your custom lowercase function so that digits are replaced with *.


Example:
```
Input: "A1B2c3"
Output: "a*b*c*"
```

Rules:

* Uppercase ‚Üí convert to lowercase

* Lowercase ‚Üí keep

* Digits ‚Üí replace with *

* Any other character ‚Üí keep as is

````python
def replace_digits(s):
  result = ''
  for i in s:
    if ord(i) >= 65 and ord(i) <= 90:
      result += chr(ord(i)+32)
    elif ord(i) > 90 and ord(i) <= 122:
      result += i
    else:
      result += "*"
  return result

replace_digits("A1B2c3")
````

# üìå QUESTION 7 ‚Äî MINI QUIZ

(Topic: replacing spaces + whitespace handling)

**1Ô∏è‚É£ EASY**

> Write a function:
```
replace_spaces(s)
```

That replaces every space " " in the string with an underscore "_".

>Example:
```
Input: "this is fun"
Output: "this_is_fun"
```

Only replace spaces, not tabs or newlines.

````python
def replace_spaces(s):
  nw= ''
  for i in s:
    if i == " ":
      nw += "_"
    else:
      nw += i
  return nw

replace_spaces("this is fun")
````

**2Ô∏è‚É£ MEDIUM**

> Write a function:
```
replace_all_whitespace(s)
```

That replaces all types of whitespace (space, tab \t, newline \n) with underscores.

You must use **.isspace().**

>Example:
```
Input: "hey\tbro how\nare you"
Output: "hey_bro_how_are_you"
```

````python
def replace_all_whitespace(s):
  nw= ''
  for i in s:
    if i.isspace():
      nw += "_"
    else:
      nw += i
  return nw

replace_spaces("hey\tbro how\nare you")
````
